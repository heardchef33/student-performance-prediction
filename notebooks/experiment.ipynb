```python
y = processed_final['output_grade']

X = processed_final[processed_final.columns[:-1]]
```

```python
from sklearn.model_selection import train_test_split

X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)

X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.2222222222, random_state=42, stratify=y_train_val)
```

```python
ORDINAL_COLUMNS = X_train.select_dtypes(include=np.number).columns.to_list()

CATEGORICAL_COLUMNS = X_train.select_dtypes(include=('object')).columns.to_list()
```

```python
CATEGORICAL_COLUMNS
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(drop='first', handle_unknown='ignore')

enc.fit(X_train[CATEGORICAL_COLUMNS])

ENCODED_COLUMNS = list(enc.get_feature_names_out(CATEGORICAL_COLUMNS))
```

```python
X_train[ENCODED_COLUMNS] = enc.transform(X_train[CATEGORICAL_COLUMNS]).toarray()

X_val[ENCODED_COLUMNS] = enc.transform(X_val[CATEGORICAL_COLUMNS]).toarray()
```

```python
X_test[ENCODED_COLUMNS] = enc.transform(X_test[CATEGORICAL_COLUMNS]).toarray()
```

```python
X_train_final = X_train.copy()[ORDINAL_COLUMNS + ENCODED_COLUMNS]

X_val_final = X_val.copy()[ORDINAL_COLUMNS + ENCODED_COLUMNS]

X_test_final = X_test.copy()[ORDINAL_COLUMNS + ENCODED_COLUMNS]
```

```python
from sklearn.metrics import confusion_matrix, roc_auc_score, f1_score

def train_and_validate(model, X_train=X_train_final, X_val=X_val_final, y_train=y_train, y_val=y_val): 
    model.fit(X_train, y_train)
    y_train_preds = model.predict(X_train)
    y_train_prob = model.predict_proba(X_train)
    y_val_preds = model.predict(X_val)
    y_val_prob = model.predict_proba(X_val)
    # Training metrics  
    print('Training Metrics')
    print(f'F1 Score:{f1_score(y_train, y_train_preds, average='weighted')}')

    print(f'{confusion_matrix(y_train, y_train_preds, normalize='true')}')

    print(f'ROC:{roc_auc_score(y_train, y_train_prob, multi_class='ovr')}')


    print('--' * 30)

    # Validation metrics 

    print('Validation Metrics')
    print(f'F1 Score:{f1_score(y_val, y_val_preds, average='weighted')}')

    print(f'{confusion_matrix(y_val, y_val_preds, normalize='true')}')

    print(f'ROC:{roc_auc_score(y_val, y_val_prob, multi_class='ovr')}')
```

```python
from sklearn.ensemble import GradientBoostingClassifier

bgc = GradientBoostingClassifier(learning_rate=0.011563640674119394, max_depth=1, n_estimators=158)

train_and_validate(bgc)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()

train_and_validate(rfc)
```

```python

def ROC(model, X_train=X_train_final, X_val=X_val_final, y_train=y_train, y_val=y_val): 
    
    model.fit(X_train, y_train)

    y_val_prob = model.predict_proba(X_val)

    return roc_auc_score(y_val, y_val_prob, multi_class='ovr')
```

```python
import optuna

from optuna import Trial

def objective(trial: Trial): 

    n_estimators = trial.suggest_int('n_estimators', 300, 400)
    
    max_depth = trial.suggest_int('max_depth', 50, 70)
    
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 4)

    min_sample_split = trial.suggest_int('min_samples_split',2, 4)

    model = RandomForestClassifier(

        n_estimators=n_estimators,

        max_depth=max_depth,

        min_samples_leaf=min_samples_leaf,

        min_samples_split=min_sample_split,

        random_state=42

    )

    score = ROC(model)

    return score

    
```

```python
study = optuna.create_study(direction='maximize')
```

```python
study.optimize(objective, n_trials=40)
```

```python
study.best_params
```

```python
def test(model, X_train=X_train_final, X_test=X_test_final, y_train=y_train, y_test=y_test): 
    
    model.fit(X_train, y_train)

    y_test_prob = model.predict_proba(X_test)

    return roc_auc_score(y_test, y_test_prob, multi_class='ovr')
```

```python
## FINAL MODEL 

final_rfc = RandomForestClassifier(
    n_estimators=381,
    max_depth=52,
    min_samples_leaf=1,
    min_samples_split=4
)

train_and_validate(final_rfc)

```

```python
test(final_rfc)
```
